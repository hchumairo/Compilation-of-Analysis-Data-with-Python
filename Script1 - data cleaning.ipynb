```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
df = pd.read_csv(pwd + "\\data_cleaning_challenge.csv")
```

```python
df1 = df.copy()
df1
```

```python
df1 = df1.drop(columns = ["Unnamed: 9", "Unnamed: 10"])
df1
```

```python
df1 = df1[df1["Row Type"].notna()]
df1
```

```python
column_values = []
counter = 0

for i in df1["Row Type"]:
    if "first name" in i:
        counter += 1
    column_values.append(counter)
```

```python
column_values
```

```python
iter_cols = df1
iter_cols["Iteration"] = column_values
iter_cols
```

```python
drop_extra_column_name_rows = iter_cols[iter_cols["Row Type"] != "Row Type"]
drop_extra_column_name_rows
```

```python
name_dataframe = drop_extra_column_name_rows[drop_extra_column_name_rows["Row Type"].str.contains("first name")]
name_dataframe.drop(columns = ["Speed1", "Speed2", "Electricity", "Effort", "Weight", "Torque"], inplace=True)
name_dataframe.rename(columns = {"Row Type" : "First Name", "Iter Number" : "Last Name", "Power1" : "Date"}, inplace=True)
name_dataframe["First Name"] = name_dataframe["First Name"].str.split[12:]
name_dataframe["Last Name"] = name_dataframe["Last Name"].str[11:]
name_dataframe["Date"] = name_dataframe["Date"].str[6:]
name_dataframe
```

```python
no_name_dataframe = drop_extra_column_name_rows[~drop_extra_column_name_rows["Row Type"].str.contains("first name")]
no_name_dataframe
```

```python
df2 = pd.merge(left=name_dataframe, right=no_name_dataframe, how="inner", on="Iteration")
df2
```
